```python
import requests
import json
import pandas as pd
```

```python
url = 'https://www.govtrack.us/api/v2/role?current=true&role_type=representative&limit=438'
```

```python
r = requests.get(url)
```

```python
data = json.loads(r.text)
```

```python
print(data)
```

```python
print(len(data))
```

```python
print(json.dumps(data, indent = 3))
```

```python
print(data['meta']['limit'])
```

```python
print(data['objects'])
```

```python
for item in data:
    print(item)
```

```python
for item in data['meta']:
    print(item)
```

```python
for item in data['objects']:
    print(item)
```

```python
details = []
for item in data['objects']:
    congress_numbers = item['congress_numbers']
    description = item['description']
    enddate = item['enddate']
    role_type = item['role_type']
    office = item['extra']['office']
    firstname = item['person']['firstname']
    gender = item['person']['gender']
    nickname = item['person']['nickname']
    details_dict = {
        'congress_numbers':congress_numbers,
        'description':description,
        'enddate':enddate,
        'role_type':role_type,
        'office':office,
        'firstname':firstname,
        'gender':gender,
        'nickname':nickname
        
    }
    details.append(details_dict)
print(json.dumps(details, indent = 3))
```

```python
df = pd.DataFrame(details)
df.to_csv('list.csv', index=False)
```
