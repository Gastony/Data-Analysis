```python
import requests
import json
import pandas as pd
```

```python
url = 'https://datausa.io/api/data?drilldowns=State&measures=Population'
```

```python
r = requests.get(url)
```

```python
data = json.loads(r.text)
```

```python
print(data)
```

```python
print(len(data))
```

```python
print(json.dumps(data, indent = 3))
```

```python
for item in data:
    print(item)
```

```python
for item in data['data']:
    print(json.dumps(item, indent = 3))
```

```python
details = []
for item in data['data']:
    ID_State = item['ID State']
    State = item['State']
    ID_Year = item['ID Year']
    Year = item['Year']
    Population = item['Population']
    Slug_State = item['Slug State']
    
    details_dict = {
        'ID_State':ID_State,
        'State':State,
        'ID_Year':ID_Year,
        'Year':Year,
        'Population':Population,
        'Slug_State':Slug_State
        
        
    }
    details.append(details_dict)
print(json.dumps(details, indent = 3))
```

```python
df = pd.DataFrame(details)
df.to_csv('extracted_data.csv', index=False)
```
